### Cluster the customers of a mall with respect to the underlying similarities among them

```r
# Importing the dataset
dataset = read.csv('Mall_Customers.csv')
dataset = dataset[4:5]
```

```r
head(dataset)
```

Spending score is a parameter computed by the mall for each customer based on their salary, spending capacity and the number of visits they pay to the mall

```r
# Using the elbow method to find the optimal number of clusters
options(repr.plot.width=5, repr.plot.height=5)

set.seed(6)
wcss = vector()
for (i in 1:10) wcss[i] = sum(kmeans(dataset, i)$withinss)
plot(1:10,
     wcss,
     type = 'b',
     main = paste('The Elbow Method'),
     xlab = 'Number of clusters',
     ylab = 'WCSS')
```

It can be clearly deduced that elbow is formed at 5 implying that the approprite number of clusters is 5

```r
# Fitting K-Means to the dataset
set.seed(29)
kmeans = kmeans(x = dataset, centers = 5, iter.max = 300, nstart = 10)
y_kmeans = kmeans$cluster
```

```r
# Visualising the clusters
library(cluster)
clusplot(dataset,
         y_kmeans,
         lines = 0,
         shade = TRUE,
         color = TRUE,
         labels = 2,
         plotchar = FALSE,
         span = TRUE,
         main = paste('Clusters of customers'),
         xlab = 'Annual Income',
         ylab = 'Spending Score')
```

The five different clusters can be categorized into different categories based on the spending score and annual income. 

Cluster 1  : High Income and Low Spending Score - Careful Customers

Cluster 2  : Average Income and Average Spending Score - Standard Customers

Cluster 3  : High Income and High Spending Score - Target Customers

Cluster 4  : Low Income and Low Spending Score - Sensible Customers.

Cluster 5  : Low Income but High Spending Score - Careless Customers
