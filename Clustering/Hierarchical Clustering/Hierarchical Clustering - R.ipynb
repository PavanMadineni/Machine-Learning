### Cluster the customers of a mall with respect to the underlying similarities among them

```r
# Importing the dataset
dataset = read.csv('Mall_Customers.csv')
dataset = dataset[4:5]
```

```r
head(dataset)
```

Spending score is a parameter computed by the mall for each customer based on their salary, spending capacity and the number of visits they pay to the mall

```r
# Using the dendrogram to find the optimal number of clusters
dendrogram = hclust(d = dist(dataset, method = 'euclidean'), method = 'ward.D')
plot(dendrogram,
     main = paste('Dendrogram'),
     xlab = 'Customers',
     ylab = 'Euclidean distances')
```

If a horizontal line is drawn across the longest vertical line, it can be clearly seen that this line crosses 5 verticals thereby implying that the optimum number of clusters is 5.

```r
# Fitting Hierarchical Clustering to the dataset
hc = hclust(d = dist(dataset, method = 'euclidean'), method = 'ward.D')
y_hc = cutree(hc, 5)
```

```r
# Visualising the clusters
options(repr.plot.width=5, repr.plot.height=5)

library(cluster)
clusplot(dataset,
         y_hc,
         lines = 0,
         shade = TRUE,
         color = TRUE,
         labels= 2,
         plotchar = FALSE,
         span = TRUE,
         main = paste('Clusters of customers'),
         xlab = 'Annual Income',
         ylab = 'Spending Score')
```

The five different clusters can be categorized into different categories based on the spending score and annual income. 

Cluster 1  : High Income and High Spending Score - Target Customers

Cluster 2  : High Income and Low Spending Score - Careful Customers

Cluster 3  : Average Income and Average Spending Score - Standard Customers

Cluster 4  : Low Income and Low Spending Score - Sensible Customers.

Cluster 5  : Low Income but High Spending Score - Careless Customers
